#### The heatmap, as we can see is not giving us a good correlation of various attributes of the dataset. The purpose behind plotting his heatmap was to study the basic correlation between each column so that it might give us more ideas to do exploratory data analysis

#### Before we create our Decision tree model and Random forest model if we have to manipulate the data a little bit. We can see from the info() function that there is one column which is of string value and that we have to convert into numerical column so that we can train our model better and so it can also predit the values more accurately.

#### For this purpose we will use the pd.get_dummies() method. It will surely add many more columns (not actually many in this case) but it will be worth it as numerical columns do add to the better performance of the model and better predictions.

```python
column = ['purpose']
```

```python
loan_data = pd.get_dummies(data=df,columns=column, drop_first=True)
```

```python
loan_data.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix

dtree = DecisionTreeClassifier()
rfc = RandomForestClassifier()
```

```python
X = loan_data.drop('not.fully.paid', axis=1)
y = loan_data['not.fully.paid']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

## Decision Tree

```python
dtree.fit(X_train, y_train)
```

```python
tree_predictions = dtree.predict(X_test)
```

```python
print('Confusion Matrix')
print(confusion_matrix(y_test, tree_predictions))
print('\n')
print('Classification Report')
print(classification_report(y_test, tree_predictions))
```

## Random Forest

```python
rfc.fit(X_train, y_train)
```

```python
rfc_predictions = rfc.predict(X_test)
```

```python
print('Confusion Matrix')
print(confusion_matrix(y_test, rfc_predictions))
print('\n')
print('Classification Report')
print(classification_report(y_test, rfc_predictions))
```
